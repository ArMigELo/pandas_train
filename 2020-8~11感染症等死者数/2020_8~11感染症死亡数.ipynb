```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("input/8-16保管第9表.csv")
df2 = pd.read_csv("input/9-16保管第9表.csv")
df3 = pd.read_csv("input/10-16保管第9表.csv")
df4 = pd.read_csv("input/11-16保管第9表.csv")
df1.head()
```

```python
df1 = df1.replace("-",0)
df2 = df2.replace("-",0)
df3 = df3.replace("-",0)
df4 = df4.replace("-",0)
df1.head()
```

```python
df1 = df1.drop("場所",axis=1)
df2 = df2.drop("場所",axis=1)
df3 = df3.drop("場所",axis=1)
df4 = df4.drop("場所",axis=1)
df1.head()
```

```python
df1 = df1.astype(int)
df2 = df2.astype(int)
df3 = df3.astype(int)
df4 = df4.astype(int)
```

```python
# df = pd.merge(df1,df2,df3,df4, on = "新型コロナウイルス感染症", how="inner")
# df = pd.merge(df1,df2, on = "新型コロナウイルス感染症", how = "inner")
# df = pd.merge(df,df3, on = "新型コロナウイルス感染症", how = "inner")
# df = pd.merge(df,df4, on = "新型コロナウイルス感染症", how = "inner")

# df = pd.concat([df1,df2] , axis=1)
# df = pd.concat([df,df3] , axis=1)
# df = pd.concat([df,df4] , axis=1)

df = df1 + df2
df = df + df3
df = df + df4
df.head()
```
